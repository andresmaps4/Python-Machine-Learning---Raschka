```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Adaline
from mlxtend.plotting import plot_decision_regions
```

```python
s = (r'C:\Users\USUARIO\Desktop\Personal\Cursos y lectures\Python Machine Learning - Raschka\Datasets\iris.txt')
```

```python
df = pd.read_csv(s, header = None, encoding = "utf-8")
```

```python
# Select setosa and versicolor
y = df.iloc[:100, 4].values
y = np.where(y=="Iris-setosa", -1, 1)
```

```python
# Extract sepal lenght and petal lenght
X = df.iloc[:100, [0,2]].values 
```

### Feature scaling - Estandarizaci√≥n

```python
X_std = np.copy(X) 
```

```python
# Se transforman las dos columnas del conjunto de datos (se estandarizan)
X_std[:,0] = (X[:,0]-X[:,0].mean())/(X[:,0].std())
X_std[:,1] = (X[:,1]-X[:,1].mean())/(X[:,0].std())
```

```python
ada_sgd = Adaline.AdalineSGD(n_iter=15, eta=0.01, random_state=1)
ada_sgd.fit(X_std, y)

plot_decision_regions(X_std, y, clf=ada_sgd)
plt.title("Adaline - Stochastic Gradient Descent")
plt.xlabel("sepal length [standarized]")
plt.ylabel("petal length [standarized]")
plt.legend(loc="upper left")
plt.show()
plt.plot(range(1, len(ada_sgd.cost_) + 1), ada_sgd.cost_, marker="o")
plt.xlabel("Epochs")
plt.ylabel("Average Cost")
plt.tight_layout()
plt.show()
```
